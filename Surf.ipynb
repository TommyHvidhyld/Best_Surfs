```python
windspeed = []
winddirection = []
waveheight = []
airtemp = []
watertemp = []

url = 'https://api.stormglass.io/v2/weather/point'
for x, y in zip(surfcali6['latitude'], surfcali6['longitude']):
    try:
        lat = (f'{y}')
        lng = (f'{x}')

        params ={
            'lat': lat,
            'lng': lng,
            'params': ','.join(['waveHeight', 'airTemperature', 'windSpeed', 'windDirection', 'waterTemperature']),
            'start': start.to('UTC').timestamp(),  # Convert to UTC timestamp
            'end': end.to('UTC').timestamp()  # Convert to UTC timestamp
        }
        headers ={
        'Authorization': 'd87dfa70-985c-11ed-b59d-0242ac130002-d87dfade-985c-11ed-b59d-0242ac130002'
        }

        response = requests.get(url, params=params, headers=headers)
        json_data3 = response.json()

        windspeed_data = json_data3['hours'][1]['windSpeed']['icon']
        winddirection_data = json_data3['hours'][1]['windDirection']['icon']
        waveheight_data = json_data3['hours'][1]['waveHeight']['icon']
        airtemp_data = json_data3['hours'][1]['airTemperature']['noaa']
        watertemp_data = json_data3['hours'][1]['waterTemperature']['meto']

        windspeed.append(windspeed_data)
        winddirection.append(winddirection_data)
        waveheight.append(waveheight_data)
        airtemp.append(airtemp_data)
        watertemp.append(watertemp_data)

    except:
        print('failed')
        print(f'at {x},{y}')
```

```python
df = pd.DataFrame({"windspeed":windspeed,
                    "winddirection":winddirection,
                    "waveheight":waveheight,
                    "airtemp":airtemp,
                    "watertemp":watertemp

})
```

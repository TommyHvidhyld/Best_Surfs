```python
windspeed = []
winddirection = []
waveheight = []
airtemp = []
watertemp = []
cloudcover = []
gust = []
precipitation = []
visibility = []
wavedirection = []

url = 'https://api.stormglass.io/v2/weather/point'
for x, y in zip(surfcali1['latitude'], surfcali1['longitude']):
    try:
        lat = (f'{y}')
        lng = (f'{x}')

        params ={
            'lat': lat,
            'lng': lng,
            'params': ','.join(['waveHeight', 'airTemperature', 'windSpeed', 'windDirection', 'waterTemperature','cloudCover','gust','precipitation','visibility','waveDirection']),
            'start': start.to('UTC').timestamp(),  # Convert to UTC timestamp
            'end': end.to('UTC').timestamp()  # Convert to UTC timestamp
        }
        headers ={
        'Authorization': 'd87dfa70-985c-11ed-b59d-0242ac130002-d87dfade-985c-11ed-b59d-0242ac130002'
        }

        response = requests.get(url, params=params, headers=headers)
        json_data3 = response.json()

        windspeed_data = json_data3['hours'][0]['windSpeed']['sg']
        winddirection_data = json_data3['hours'][0]['windDirection']['sg']
        waveheight_data = json_data3['hours'][0]['waveHeight']['sg']
        airtemp_data = json_data3['hours'][0]['airTemperature']['sg']
        watertemp_data = json_data3['hours'][0]['waterTemperature']['sg']
        cloudcover_data = json_data3['hours'][0]['cloudCover']['sg']
        gust_data = json_data3['hours'][0]['gust']['sg']
        precipitation_data = json_data3['hours'][0]['precipitation']['sg']
        visibility_data = json_data3['hours'][0]['visibility']['sg']
        wavedirection_data = json_data3['hours'][0]['waveDirection']['sg']

        windspeed.append(windspeed_data)
        winddirection.append(winddirection_data)
        waveheight.append(waveheight_data)
        airtemp.append(airtemp_data)
        watertemp.append(watertemp_data)
        cloudcover.append(cloudcover_data)
        gust.append(gust_data)
        precipitation.append(precipitation_data)
        visibility.append(visibility_data)
        wavedirection.append(wavedirection_data)

    except:
        print('failed')
        print(f'at {x},{y}')
```
